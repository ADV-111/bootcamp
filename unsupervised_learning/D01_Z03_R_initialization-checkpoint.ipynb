# k-means - inicjalizacja

Algorytm zależy od początkowej inicjalizacji (funkcja kosztu jest niewypukła). 

## Zad. 
Proszę uruchomić algorytm na danych Iris 100 razy z różnymi inicjalizacjami:

  * random
  * k-means++

Zaraportować błąd średniokwadratowy oraz zwizualizować wyniki.

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.4, 
                  shuffle=True, 
                  random_state=0)

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
k=3

SSE_array_random = []
SSE_array_pp = []
for i in range(1000):
    km_random = KMeans(n_clusters=k, n_init=1, init='random')
    km_pp = KMeans(n_clusters=k, n_init=1, init='k-means++')

    y_km_random = km_random.fit_predict(X)
    y_km_pp = km_pp.fit_predict(X)
    
    SSE_array_random.append(km_random.inertia_)
    SSE_array_pp.append(km_pp.inertia_)
    
#     plt.scatter(X[:, 0], X[:, 1], c=y_km)
#     plt.show()
    
# plt.legend()
# plt.show()
```

```python
plt.plot( SSE_array_random, label="Random initialization", alpha = 0.5)
# plt.plot( SSE_array_pp, label="k-means++ initialization", alpha = 1 )
# plt.ylim(40,60)
plt.legend()
plt.show()
```
